```python
import numpy as np
import pandas as pd
```

```python
nineoneone_df = pd.read_csv("../resources/Call_Data.csv")
```

```python
nineoneone_df.head(5)
```

```python
nineoneone_df['Original Time Queued'] =  pd.to_datetime(nineoneone_df['Original Time Queued'], format="%m/%d/%Y %H:%M:%S %p")#, errors="ignore")
nineoneone_df['Arrived Time'] =  pd.to_datetime(nineoneone_df['Arrived Time'], format="%b %d %Y %H:%M:%S:%f%p")#, errors="ignore")
nineoneone_df["time_to_arrival"] = nineoneone_df["Arrived Time"] - nineoneone_df["Original Time Queued"]
nineoneone_df["time_to_arrival"] = nineoneone_df["time_to_arrival"].dt.seconds.div(60)
nineoneone_df['call_hour'] = nineoneone_df['Original Time Queued'].dt.hour
# nineoneone_df[[ "Original Time Queued", "Arrived Time", "time_to_arrival", 'call_hour']]
nineoneone_df.head()
```

```python
nineoneone_df.plot.scatter(x='call_hour',
                      y='time_to_arrival',
#                       colormap='viridis'
                          )
```

```python
nineoneone_df.plot.scatter(x='Final Call Type',
                      y='time_to_arrival',
#                       colormap='viridis',
                        figsize=(15,8)
                          )
```

```python
noon_response_times = nineoneone_df["time_to_arrival"]
noon_response_times = nineoneone_df[nineoneone_df['call_hour']==12]
boxplot = noon_response_times.boxplot(column=['time_to_arrival'])
# noon_response_times
```
